```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def func(x, tn, t, g):
    return g*(tn + t*(x + 0.00000001)/np.tanh(x + 0.00000001))
```

```python
xdata = np.linspace(-10, 10, 99)
y = func(xdata, 2.5, 2, 1)
np.random.seed(1729)
y_noise = 0.4 * np.random.normal(size=xdata.size)
ydata = y + y_noise
plt.plot(xdata, ydata, 'b+', label='data')
```

```python
popt, pcov = curve_fit(func, xdata, ydata)
```

```python
popt
```

```python
plt.plot(xdata,ydata-func(xdata, *popt),'g+',xdata,ydata,'b+',xdata, func(xdata, *popt),'r-',
...          label='fit: a=%5.3f, b=%5.3f, c=%5.3f' % tuple(popt))
```

```python
def func2(x, tn, t, g,f):
    xplus = (x + 0.0000000001 + f)/(2*t)
    xminus = (x + 0.0000000001 - f)/(2*t)
    return g*(tn + t*(xplus/np.tanh(xplus)) + t*(xminus/np.tanh(xminus)))
```

```python
xdata = np.linspace(-20, 20, 99)
y = func2(xdata, 1, 0.5, 1, 6)
np.random.seed(1729)
y_noise = 0.2 * np.random.normal(size=xdata.size)
ydata = y + y_noise
plt.plot(xdata, ydata, 'b+', label='data')
```

```python
popt, pcov = curve_fit(func2, xdata, ydata)
```

```python
plt.plot(xdata,ydata-func2(xdata, *popt),'g+',xdata,ydata,'b+',xdata, func2(xdata, *popt),'r-')
```
